```python
##13. Apply K-Means method.

#Using PCA to get two vectors out of 10 variables from the dataframe
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
#scaling data
scaler=StandardScaler()
scaled_data=scaler.fit_transform(Clients_df)
scaled_data
```

```python
from sklearn.decomposition import PCA
n_components=10
pca=PCA(n_components=n_components)
principal_components=pca.fit_transform(scaled_data)
```

```python
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
cumulative_explained_variance = explained_variance.cumsum()
plt.plot(range(1, n_components + 1), cumulative_explained_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
selected_components = 2
final_data = principal_components[:, :selected_components]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8)
Clients_df['Cluster'] = kmeans.fit_predict(final_data)
```

```python
import seaborn as sns

## 14. Visualize clusters and summarize your findings.
Clients_df_pca = pd.DataFrame(final_data, columns=['PC1', 'PC2'])
Clients_df_pca['Cluster'] = Clients_df['Cluster']

sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=Clients_df_pca, palette='viridis')
plt.title('Scatterplot of Transformed Data (PCA)')
plt.show()
```
